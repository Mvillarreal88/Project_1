```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# pull in data from csv

us_df = pd.read_csv("Data_csv/US.csv")
us_obes_df = pd.read_csv("Data_csv/US_obes.csv")
us_Pa_df = pd.read_csv("Data_csv/US_PA.csv")
```

```python
gen_com = us_df[['Difference in male life expectancy, 1985-2010 (years)', 'Difference in female life expectancy, 1985-2010 (years)']]
gen_com = gen_com.transpose()
gen_com = gen_com.rename(index={'Difference in male life expectancy, 1985-2010 (years)': 'Male', 'Difference in female life expectancy, 1985-2010 (years)': 'Female'})
gen_com.plot(kind='bar', color='#FDD835', legend=False)

# # Set Title and labels
plt.title('Difference in life expectancy, 1985-2010 (years)')
plt.xlabel('Gender')
plt.ylabel('Years')
plt.savefig('output_data/Difference in life expectancy, 1985-2010 (years).png', bbox_inches='tight')
plt.show()
```

```python
gen_com_obes = us_obes_df[['Difference in male obesity prevalence, 2001-2009 (percentage points)', 'Difference in female obesity prevalence, 2001-2009 (percentage points)']]
gen_com_obes = gen_com_obes.transpose()
gen_com_obes = gen_com_obes.rename(index={'Difference in male obesity prevalence, 2001-2009 (percentage points)': 'Male', 'Difference in female obesity prevalence, 2001-2009 (percentage points)': 'Female'})
gen_com_obes.plot(kind='bar', color='#3949AB', legend=False)

# # Set Title and labels
plt.title('Difference in obesity prevalence, 2001-2009 (percentage points)')
plt.xlabel('Gender')
plt.ylabel('Change in Percentage')
plt.savefig('output_data/Difference in obesity prevalence, 2001-2009 (percentage points).png', bbox_inches='tight')
plt.show()
```

```python
us_Pa_df

gen_com_Pa = us_Pa_df[['Difference in male sufficient physical activity prevalence, 2001-2009 (percentage points)', 'Difference in female sufficient physical activity prevalence, 2001-2009 (percentage points)']]
gen_com_Pa = gen_com_Pa.transpose()
gen_com_Pa = gen_com_Pa.rename(index={'Difference in male sufficient physical activity prevalence, 2001-2009 (percentage points)': 'Male', 'Difference in female sufficient physical activity prevalence, 2001-2009 (percentage points)': 'Female'})
gen_com_Pa.plot(kind='bar', color='#E53935', legend=False)

# # Set Title and labels
plt.title('Difference in sufficient physical activity prevalence, 2001-2009 (percentage points)')
plt.xlabel('Gender')
plt.ylabel('Change in Percentage')
plt.savefig('output_data/Difference in sufficient physical activity prevalence, 2001-2009 (percentage points).png', bbox_inches='tight')
plt.show()
```

```python
x = np.arange(6)
# y1985 = [us_df['Male life expectancy, 1985 (years)'][0], us_df['Female life expectancy, 1985 (years)'][0]]
# y1990 = [us_df['Male life expectancy, 1990 (years)'][0], us_df['Female life expectancy, 1990 (years)'][0]]
# y1995 = [us_df['Male life expectancy, 1995 (years)'][0], us_df['Female life expectancy, 1995 (years)'][0]]
# y2000 = [us_df['Male life expectancy, 2000 (years)'][0], us_df['Female life expectancy, 2000 (years)'][0]]
# y2005 = [us_df['Male life expectancy, 2005 (years)'][0], us_df['Female life expectancy, 2005 (years)'][0]]
# y2010 = [us_df['Male life expectancy, 2010 (years)'][0], us_df['Female life expectancy, 2010 (years)'][0]]
male_life_ex = [us_df['Male life expectancy, 1985 (years)'][0],
                us_df['Male life expectancy, 1990 (years)'][0],
                us_df['Male life expectancy, 1995 (years)'][0],
                us_df['Male life expectancy, 2000 (years)'][0],
                us_df['Male life expectancy, 2005 (years)'][0],
                us_df['Male life expectancy, 2010 (years)'][0]]
female_life_ex = [us_df['Female life expectancy, 1985 (years)'][0],
                us_df['Female life expectancy, 1990 (years)'][0],
                us_df['Female life expectancy, 1995 (years)'][0],
                us_df['Female life expectancy, 2000 (years)'][0],
                us_df['Female life expectancy, 2005 (years)'][0],
                us_df['Female life expectancy, 2010 (years)'][0]]
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, male_life_ex, width, color='lightblue')
plt.bar(x, female_life_ex, width, color='orange')
# # Set Title and labels
plt.title('Male and Female Life Expectancy (1985 - 2010)')
plt.xlabel('Year')
plt.ylabel('Age')
# plt.xlim(-0.25, len(x_axis))
plt.ylim(65, 82)
x_axis = np.arange(len(male_life_ex))
Gender=['Male','Female']
labels = [1985, 1990, 1995, 2000, 2005, 2010]
tick_locations = [value-0.1 for value in x_axis]
plt.xticks(tick_locations, labels, rotation="vertical")
plt.legend(Gender,loc=2)
plt.savefig('output_data/Male and Female Life Expectancy (1985 - 2010).png', bbox_inches='tight')
plt.show()
```
