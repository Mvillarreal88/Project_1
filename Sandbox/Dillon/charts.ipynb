```python
state_data_df = state_data_df.sort_values('Difference in male life expectancy, 1985-2010 (years)', ascending=False)
# set x axis and tick locations
x_axis = np.arange(len(state_data_df))
tick_locations = [value+0.4 for value in x_axis]
# create bar chart and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, 
        state_data_df["Difference in male life expectancy, 1985-2010 (years)"], 
        color='#FFB300', 
        align="edge")
plt.xticks(tick_locations, state_data_df["State"], rotation="vertical")
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(-0, 8)
# Set Title and labels
plt.title("Difference in male life expectancy, 1985-2010 (years)")
plt.xlabel("State")
plt.ylabel("Years")
plt.show()
```

```python
gen_com = us_df[['Difference in male life expectancy, 1985-2010 (years)', 'Difference in female life expectancy, 1985-2010 (years)']]
gen_com = gen_com.transpose()
gen_com = gen_com.rename(index={'Difference in male life expectancy, 1985-2010 (years)': 'Male', 'Difference in female life expectancy, 1985-2010 (years)': 'female'})
gen_com.plot(kind='bar', color='#FDD835', legend=False)

# # Set Title and labels
plt.title('Difference in life expectancy, 1985-2010 (years)')
plt.xlabel('Gender')
plt.ylabel('Years')
plt.show()
```
